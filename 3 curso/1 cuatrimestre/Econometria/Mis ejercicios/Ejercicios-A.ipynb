### Pregunta 6: (1.5 puntos)

Se desea analizar el modelo $Y_ t = \beta_0 + \beta_1 X_{1t} + \beta_2 X_{2t} + u_t$, con 
        $$(\mathbf{X}^{t} \mathbf{X}) =  \begin{pmatrix} 14 & 7 & 14\\ 7 & 4.5 & 7\\ 14 & 7 & 15 \end{pmatrix}, \mathbf{X}^{t} \mathbf{y} = 
            \begin{pmatrix}
                10\\6\\12
            \end{pmatrix}, \mathbf{y}^t \mathbf{y} =14$$
            
Se pide responder de forma razonada las siguientes cuestiones:
1. Obtener la estimación mínimo cuadrática de los coeficientes del modelo.
1. Verificar la signifaicación del modelo.
1. Calcular la predicción puntual para el valor individual, así como el intervalo de predicción para este valor al $95\%$ sabiendo que $x_1=5$ y $x_2=7$.

```python
#Cálculos:

#Cálculos:
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pylab as plt
from wooldridge import *
import statsmodels.stats.outliers_influence as oi #Para el test RESET de Ramsey
import statsmodels.stats.diagnostic as diagn #Para el test de Harvey-Collier
import pandas as pd
from scipy import stats

X_trasp_X = np.array([[14,7,14],[7,4.5,7],[14,7,15]])

X_trasp_X_inversa = np.linalg.inv(X_trasp_X)

X_trasp_y = np.array([[10], [6], [12]])
y_trasp_y = 14

Beta_hat = np.dot(X_trasp_X_inversa, X_trasp_y)
print("Nuestro beta hat: ", Beta_hat)

n = X_trasp_X[0][0]
k = 2
print("Numero de muestras: ", n)
print("Numero de variables: ", k)

scr = y_trasp_y - np.dot(np.transpose(Beta_hat), X_trasp_y)
sce = np.dot(np.transpose(Beta_hat), X_trasp_y) - np.dot(n, (X_trasp_y[0]/n)**2)

alpha = 0.05

F_exp = ((n-k)/(k-1))*(sce/scr)
F_teorica = stats.f.ppf((1-alpha), (k-1), (n-k))

print("\nF teorica: ", F_teorica)
print("F experimental: ", F_exp)
print("Como nuestra f-experimental es mayor que la f-teorica, se rechaza la hipotesis nula y decimos que tenemos significacion global")

#------------------------------------------
# Apartado 3

sigma_hat_sq = scr/(n-k)
matriz_varianza_covarianza = sigma_hat_sq * X_trasp_X_inversa

t_exp_1 = np.absolute(Beta_hat[1]/np.sqrt(matriz_varianza_covarianza[1,1]))
t_exp_2 = np.absolute(Beta_hat[2]/np.sqrt(matriz_varianza_covarianza[2,2]))
t_teorica = stats.t.ppf(1-(alpha/2), n-k)

print("\nPara nuestra beta_hat[1]: ")
print("T teorica: ", t_teorica)
print("T experimental: ", t_exp_1)


print("\nPara nuestra beta_hat[2]: ")
print("T teorica: ", t_teorica)
print("T experimental: ", t_exp_2)

Y = Beta_hat[0] + Beta_hat[1]*5 + Beta_hat[2]*7
print("\nNuestra prediccion es: ", Y)

```

Estima el modelo lineal utilizando los datos de *datos1.xls* y estudia la existencia de autocorrelación en éste. En caso afirmativo, aplica alguna estrategia para corregirla y comprueba si ha concluido con éxito.

```python
#Cálculos:

import pandas as pd

datos8 = pd.read_excel ("datos1.xls")

Y = datos8["inflacion"]
X_sqr = datos8["CrecimientoSalarios"]**2

mco8 = sm.OLS(Y, sm.add_constant(X_sqr)).fit()

mco8.summary()
```

```python
#Cálculos:
from statsmodels.stats.stattools import durbin_watson #Si DW es 0, correlacion positiva, si es 4, correlacion negativa y si es 2 es incorrelaccionado, si no es alguno de esos valores me fijo en las tablas

DW = durbin_watson(mco8.resid) #con el DW 0.066 tengo un problema de correlacion positiva
print("DW: ", DW, "--> Problema de autocorrelacion")
```

```python
rho= 1 - DW/2 # ro estimado inicial

print (rho) # rho inicial
mco_autocorr=sm.GLSAR(Y, sm.add_constant(X_sqr), rho=rho) # minimos cuadrados generalizados
res = mco_autocorr.iterative_fit(maxiter=1000, rtol = 10**(-10)) # iteraciones y tolerancia, la tolerancia es si entre un rho y otro hay muy poca diferencia paro (10^-10)

print ('Iterations used = %d Converged %s' % (res.iter, res.converged) )
print ('Rho = ', mco_autocorr.rho)
res.summary()
```

En la base de datos "crime2" de Wooldridge se facilita información sobre la tasa de crimen (crmrte) en diversas ciudades en los años 1982 y 1987 (variable year). En particular, estamos intereaados en utilizar esta base de datos para  saber como afecta el desempleo en la tasa de delincuencia.

```python
#Cálculos:

datos1982 = datos10[datos10["year"] == 82]

Y=datos1982["crmrte"]
X=datos1982[["unem", "crimes", "officers", "pop", "area"]]

mco10_1982 = sm.OLS(Y, sm.add_constant(X)).fit()

mco10_1982.summary()
```

```python
scr1982=mco10_1982.ssr
scr1987=mco10_1987.ssr
scr = mco10.ssr
n = mco10.nobs

k=3

alpha=0.05
Fteo= stats.f.ppf(1-alpha,k+1,n-2*k-1)

#Test de chow
Fexp = ((n-(2*k)-2)/(k-1)) * ((scr-(scr1987+scr1982))/(scr1987+scr1982))

print("La F-teorico: ", Fteo)
print("La F-experimental: ", Fexp)

print("Como nuestra f-experimental es mas grande que f-teorico, rechazamos la hipotesis nula, por lo que podemos decir que no coincide las betas de 1982 y 1987")
```
