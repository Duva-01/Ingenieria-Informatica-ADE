1. Usando los datos de **sleep75**, analizar la posible existencia de heteroscedasticidad en el modelo $
sleep = \beta_0 + \beta_1 totwrk + \beta_2 educ + \beta_3 age + \beta_4 age^2 + \beta_5 c + u$

```python
datos = dataWoo('sleep75')
y=datos["sleep"]
X=datos[["totwrk","educ","age","agesq","yngkid"]] # no se cuales son las variables

mco = sm.OLS(y, sm.add_constant(X)).fit()
mco.summary()
```

```python
import statsmodels.stats.api as sms


GQ = sms.het_goldfeldquandt(mco.resid,mco.model.exog,split = 237) #Para dividir entre 3 las variables, para comparar los maximos y minimos, tampoco tengo muy claro porque 4

print("Test de goldfeldquandt: ")
print("Descripcion --> split es el numero de elementos que pone en cada trozo, si no pones nada divide en 2")
print(GQ, "--> No tengo problema de heterocedasticidad, ya que mi p-valor > 0.05 \n")

#BREUSH-PAGAN
BP = sms.het_breuschpagan(mco.resid, mco.model.exog)

print("Test de breuschpagan: ")
print("Descripcion --> Los dos ultimos valores me interesan , el estadistico experimental y P valor")
print(BP, "--> No tengo problema de heterocedasticidad, ya que mi p-valor > 0.05 \n")



#WHITE

#El test de white es muy parecido, pero tiene en cuenta el producto de las variables
#Si me da significacion global no tengo problema de heterocedasticidad
W=sms.het_white(mco.resid,mco.model.exog)
print("Test de White: ")
print("Descripcion --> Lo mismo que los anteriores test")
print(W, "--> No tengo problema de heterocedasticidad, ya que mi p-valor > 0.05 \n")
```

3. Con la base de datos **gpa1**:
 - Estimar $colGPA = \beta_0 + \beta_1 hsGPA + \beta_2 ACT + \beta_3 skipped + \beta_4 PC + u$ y extraer los residuos.
 - Contrastar la posibilidad de heteroscedasticidad con el test de Breush-Pagan.
 - Analizar, con el test de Glejser, la dependencia funcional de la heteroscedasticidad con respecto a la variables que podrían provocarla.
 - Ajustar, según el método de Mínimos Cuadrados Ponderados, el modelo con la corrección de heteroscedasticidad.

```python
data = dataWoo('gpa1')

#Glejser

y=data["colGPA"]
X=data[["hsGPA","ACT","skipped","PC"]]

mco = sm.OLS(y, sm.add_constant(X)).fit()

print(mco.summary())
print(mco.resid)
```

```python
BP = sms.het_breuschpagan(mco.resid, mco.model.exog)
print("Test de breuschpagan: ")
print("Descripcion --> Los dos ultimos valores me interesan , el estadistico experimental y P valor")
print(BP, "--> Tengo problema de heterocedasticidad, ya que mi p-valor < 0.05 \n")
```

```python
#z=data.dropna(subset=['hsGPA'])
z=data["hsGPA"]
for h in [-2,-1,-0.5,0.5,1,2]:
    mcoaux=sm.OLS(abs(mco.resid),sm.add_constant(z**h)).fit()
    pval=mcoaux.pvalues[1]
    print("h: ",h,"-> pvalt:",pval,"R2: ", mcoaux.rsquared)
    
#nos fijamos en los pvalores mas pequenios que 0.05, y dentro de eso los que tienen el R cuadrado mas grande, 
#si lo rechazo tengo un problema, de esas cual es la que tiene el R cuadrado mas grande, Z^2, se corrige con minimos 
#cuadrados ponderados, multiplicando mis datos por 1/sqrt(Z^2)

mcp = sm.WLS(y, sm.add_constant(X),weights=1./np.sqrt (z**2)).fit() # minimos cuadrados ponderados
mcp.summary()
```

4. Usando los datos de **barium**, estima el modelo $log(chempi)= \beta_0 + \beta_1 log(gas) + \beta_2 log(rtwex) + \beta_3 befile6  + \beta_4 affile6 + \beta_5  afdec6 + u$, y analiza la posible existencia de autocorrelación en el modelo. En caso afirmativo, estima el modelo corregido.

```python
dataWoo('barium', description=True)
datos=dataWoo('barium')
y=datos['lchempi']
X=datos[['lgas', 'befile6', 'affile6', 'afdec6']]
mco4=sm.OLS(y, sm.add_constant(X)).fit()
mco4.summary()
```

```python
from statsmodels.stats.stattools import durbin_watson #Si DW es 0, correlacion positiva, si es 4, correlacion negativa y si es 2 es incorrelaccionado, si no es alguno de esos valores me fijo en las tablas
dw=durbin_watson(mco4.resid) #con el DW 0.066 tengo un problema de correlacion positiva

print ("Nuestro durbin-watson: ", dw, "--> No se aproxima a 2 por lo que tenemos un problema de autocerrelación positiva")
print("--> Si da entre los valores de 2 y resulta no concluyente, utilizamos test de Ljung-box")
print("--> Si nuestro valor da entre la autocorrelacion positiva o negativa, aplicamos el test de rho.")
```

```python
rho= 1 - dw/2 # ro estimado inicial

print (rho) # rho inicial
mco_autocorr=sm.GLSAR(y, sm.add_constant(X), rho=rho) # minimos cuadrados generalizados
res = mco_autocorr.iterative_fit(maxiter=2000, rtol = 100**(-10)) # iteraciones y tolerancia, la tolerancia es si entre un rho y otro hay muy poca diferencia paro (10^-10)

print ('Iterations used = %d Converged %s' % (res.iter, res.converged) )
print ('Rho = ', mco_autocorr.rho)
res.summary()
```

```python
dataWoo('prminwge', description=True)
```

```python
datos=dataWoo('prminwge')

y=datos['lprepop']
X=datos[['lmincov', 'lusgnp']]
        
mco5=sm.OLS(y, sm.add_constant(X)).fit()
mco5.summary() # tenemos correlacion positiva otra vez
```

```python
from statsmodels.stats.diagnostic import acorr_ljungbox 
print("Ljung-box")
print("--> Analizamos con Ljung-box si existen relaccion en mi pertubacion aleatoria con la de m años antes")
print("--> Lags son los retardos, comparo hasta con 3 años antes")
print("--> Nos da el p valor, que es la siguiente lista, rechazo las 3 hipotesis ya que todos los p valores son extremadamente chicos.")
print("--> Es decir existe relacion con los otros años")

acorr_ljungbox(mco5.resid, lags=3, return_df=True)
```

```python
from statsmodels.stats.stattools import durbin_watson

rho= 1 - dw/2 # ro estimado inicial

print (rho) # rho inicial
mco_autocorr=sm.GLSAR(y, sm.add_constant(X), rho=rho)
res = mco_autocorr.iterative_fit(maxiter=1000, rtol = 10**(-10)) # iteraciones y tolerancia, la tolerancia es si entre un rho y otro hay muy poca diferencia paro (10^-10)

print ('Iterations used = %d Converged %s' % (res.iter, res.converged) )
print ('Rho = ', mco_autocorr.rho)
res.summary()
```
